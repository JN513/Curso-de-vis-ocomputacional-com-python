## Imports

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

print(cv2.__version__)
print(np.__version__)
```

## Transformações em imagens (mudança de canais de cores)

```
path = "/content/Curso-de-visaocomputacional-com-python/imagens/imgs/8.jpg"

imagem = cv2.imread(path)
```

```
cv2_imshow(imagem)

plt.imshow(imagem)
```

Convertendo de BGR para RGB

```
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_rgb)
```

Convertendo de BGR para Gray e RGB para Gray

```
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(imagem_gray)
```

```
plt.imshow(imagem_gray, cmap='gray')
```

```
imagem_rgb.shape
```

```
imagem_gray.shape
```

## Região de interesse

```
imagem_rosto = imagem_rgb[100:200, 100:200]

plt.imshow(imagem_rosto)
```

## Salvando imagens

```
cv2.imwrite("rosto.png", imagem_rosto)
```

```
imagem_rosto_bgr = cv2.cvtColor(imagem_rosto, cv2.COLOR_RGB2BGR)

cv2.imwrite("rosto_bgr.png", imagem_rosto_bgr)
```
